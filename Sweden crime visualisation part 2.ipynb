### Crimes that are shockingly growing.

```python
#omport packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Desktop/datasets_851_1557_reported.csv')
df
```

```python
missing_val_count_by_column = (df.isnull().sum())
print(missing_val_count_by_column[missing_val_count_by_column>0])
```

```python
df = df.dropna(axis = 0)
df.head()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,3,1)
plt.plot(df['Year'],df['sexual.offenses'],c="red")
plt.grid(color="grey")
plt.xlabel('Year')
plt.ylabel('Sexual offenses')
plt.title('Sexual offenses by year')
plt.subplot(2,3,2)
plt.plot(df['Year'],df['fraud'],c="red")
plt.grid(color="grey")
plt.xlabel('Year')
plt.ylabel('Fraud')
plt.title('Fraud by year')
plt.subplot(2,3,3)
plt.plot(df['Year'],df['other.penal.crimes'],c="red")
plt.grid(color="grey")
plt.xlabel('Year')
plt.ylabel('Other penal crimes')
plt.title('Other penal crimes by year')
plt.subplot(2,3,4)
plt.plot(df['Year'],df['narcotics'],c="red")
plt.grid(color="grey")
plt.xlabel('Year')
plt.ylabel('Narcotics')
plt.title('Narcotics cases by year')
plt.subplot(2,3,5)
plt.plot(df['Year'],df['drunk.driving'],c="red")
plt.grid(color="grey")
plt.xlabel('Year')
plt.ylabel('Drunk driving')
plt.title('Drunk driving cases by year')
plt.show()
```
